```python
#Importing needed algorithms for doing analysis

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import Lasso,Ridge,ElasticNet

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
#Using a for loop for finding the best accuracy & best random state 

maxAcc=0
maxRS=0
for i in range(1,200):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=i)
    Ln=LinearRegression()
    Ln.fit(X_train,y_train)
    pred=Ln.predict(X_test)
    acc=r2_score(y_test,pred)
    if acc>maxAcc:
        maxAcc=acc
        maxRS=i
print("Best accuracy is ",maxAcc, " at Random State ",maxRS)
```

```python
#Sending the data for the training & testing phase

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=maxRS)
```

```python
#Storing all the models in a variable & using for loop to find out which one is the best

model=[LinearRegression(),DecisionTreeRegressor(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    m.fit(X_train,y_train)
    predm=m.predict(X_test)
    acc=r2_score(y_test,predm)
    print('Accuracy Score of',m,'is:',acc)
    print('mean_absolute_error:',mean_absolute_error(y_test,predm))
    print('mean_squared_error:',mean_squared_error(y_test,predm))
    print('Root mean_squared_error:',np.sqrt(mean_squared_error(y_test,predm)))
    print("\n")
```

# CROSS VALIDATION

```python
#Cross validating all the models to find which is the best model available for this dataset 

model=[LinearRegression(),DecisionTreeRegressor(),Lasso(alpha=0.0001),Ridge(alpha=0.0001),ElasticNet(alpha=0.0001)]
for m in model:
    score=cross_val_score(m,X,y,cv=5)
    print("Score for",m,"is: ",score.mean())
```

Ridge is the best model for us as their is very less amount of validation is there i.e. 0.04.

# HYPER PARAMETER TUNNING

```python
#Importing GridSearchCV for hyper parameter tunning

from sklearn.model_selection import GridSearchCV
```

```python
#Setting parameters for Ridge parameters

parameters={"max_iter":[1, 5, 10],
            "alpha":[0.0001, 0.001, 0.01, 0.1, 1, 10, 100],
            "solver": ['auto','svd','cholesky','lsqr','sparse_cg','sag','saga','lbfgs']
           }
```

```python
#Creating instance of GridSearchCV

GCV=GridSearchCV(Ridge(),parameters,cv=5)
```

```python
#Finding the best parameters among the available 

GCV.get_params()
```

```python
#Passing the best parameters for tunning

Final_mod= Ridge(alpha=1.0,max_iter=None,random_state=None)
Final_mod.fit(X_train,y_train)
pred=Final_mod.predict(X_test)
r2=r2_score(y_test,pred)
print(r2*100)
```

After hyper-parameter tunning we are getting 86 % of accuracy
