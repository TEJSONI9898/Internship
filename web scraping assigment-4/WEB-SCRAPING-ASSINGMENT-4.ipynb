```python
#Importing other needed libraries

import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
```

```python
#Creating an instance of the driver

driver = webdriver.Chrome(r'C:/Users/Deep/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
#Going to the page

driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#Creating empty lists for storing all the data in one go

Rank=[]
Name=[]
Artist=[]
Upload_d=[]
Views=[]
```

```python
#Fething & storing all the elements in the lists

for i in driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']//tbody//tr//td[1]"):
    Rank.append(i.text)
for j in driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']//tbody//tr//td[2]"):
    Name.append(j.text)
for t in driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']//tbody//tr//td[3]"):
    Artist.append(t.text)
for u in driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']//tbody//tr//td[4]"):
    Upload_d.append(u.text)
for v in driver.find_elements_by_xpath("//table[@class = 'wikitable sortable jquery-tablesorter']//tbody//tr//td[5]"):
    Views.append(v.text)
```

```python
#Creating a dataframe of the fetched list

Top_views=pd.DataFrame({})
Top_views['Rank']=Rank[:30]
Top_views['Name']=Name[:30]
Top_views['Artist']=Artist[:30]
Top_views['Date of Uploading']=Upload_d[:30]
Top_views['Views']=Views[:30]
```

# Q5

```python
#Opening Chrome

driver5=webdriver.Chrome(r'C:/Users/Deep/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
#Giving the website & getting to the URL

url="https://github.com/trending"
driver5.get(url)
```

```python
#Inputting links

links=[]
try:
    for l in driver5.find_elements_by_xpath('//h1[@class="h3 lh-condensed"]/a'):
        links.append(l.get_attribute('href'))
except NoSuchElementException:
        links.append("-")
links
```

```python
#Scraping the data

Repository_title=[]
Repository_description=[]
Contributors_count=[]
Language_used=[]
a="--"

for i in links:
    driver5.get(i)
    
    #scraping the titles
    try:
        t = driver5.find_element_by_xpath('//*[@id="js-repo-pjax-container"]/div[1]/div[1]/div/h1').text
        Repository_title.append(t.replace("\n",""))
    except NoSuchElementException:
        Repository_title.append(a)

    #scraping the description
    try:
        d = titles = driver5.find_element_by_xpath('//*[@id="repo-content-pjax-container"]/div/div[2]/div[2]/div/div[1]/div/p').text
        Repository_description.append(d) 
    except NoSuchElementException:
        Repository_description.append(a)    
                        
    #scraping the company_name 
    try:
        #c = driver.find_element_by_xpath('//div[@class="flex-shrink-0 col-12 col-md-3"]/div/div[3]/div/h2/a/span').text
        c= driver5.find_element_by_xpath('//div[@class="BorderGrid BorderGrid--spacious"]/div[2]/div/h2/a/span').text
        Contributors_count.append(c) 
    except NoSuchElementException:
        Contributors_count.append(a)

        
    #scraping the Language
    try:
        l1 = driver5.find_element_by_xpath('//a[@class="d-inline-flex flex-items-center flex-nowrap Link--secondary no-underline text-small mr-3"]').text
        Language_used.append(l1.replace("\n"," - ")) 
    except NoSuchElementException:
        Language_used.append(a)
```

```python
#Creating & printing the dataframe

Github=pd.DataFrame({"Repository title":Repository_title, "Repository description":Repository_description,
                    "Contributors count":Contributors_count,"Language used":Language_used})
Github
```

```python
#Closing the driver

driver5.close()
```
