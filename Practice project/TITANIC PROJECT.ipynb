# DROPING COLUMNS WE DON'T NEED

```python
#Dropping PassengerId column
#Name has no important to build our madel

df.drop('PassengerId', axis=1, inplace = True )
df.drop('Name',axis=1, inplace=True)   
```

```python
df.head()
```

# MISSING VALUE TREATMENT

```python
#We need to treat Age Cabin & Embarked

df.isnull().sum()
```

```python
#Age is a numeric column & have outliers so we will use median to remove it

df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
#Embarked is an categorical column so we will remove outliers with mode

df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
#It is an alpha numeric column so we can't use mean,median,mode .
#We can delete it but to bring best model we won't do it
#We will use fillna

df['Cabin'] = df['Cabin'].fillna('U')
```

```python
#All the missing values has been treated

df.isnull().sum()
```

```python
#For better understanding we are using heatmap

sns.heatmap(df.isnull())
```

# REMOVING OUTLIERS

# SCALING THE DATA

```python
#Printing the shape of x & y

x.shape, y.shape
```

```python
#With the help of MinMax scaler we will scale the dataframe

from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x = pd.DataFrame(mms.fit_transform(x), columns=x.columns)
```

```python
#Printing the first five rows after scaling

x.head()
```

# MODELS

```python
#Importing all the models that need to be checked
#using for loop to get the best accuracy at best random state

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    DTC = DecisionTreeClassifier()
    DTC.fit(x_train, y_train)
    pred = DTC.predict(x_test)
    acc=accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```

```python
#Using the train_test_split for sending the data to training & testing phase

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=maxRS)
```

```python
#Printing the shape of x & y

x.shape,y.shape
```

```python
#Printing the shapes of training & testing data of x & y

x_train.shape,y_train.shape, x_test.shape,y_test.shape
```

```python
#Importing all the needed algorithms

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score, accuracy_score
```

```python
#Using DecisionTreeClassifier 

DTC = DecisionTreeClassifier()
DTC.fit(x_train,y_train)
predDTC = DTC.predict(x_test)

print(accuracy_score(y_test, predDTC))
print(confusion_matrix(y_test, predDTC))
print(classification_report(y_test,predDTC))
```

```python
#Using RandomForestClassifier

RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
predRFC = RFC.predict(x_test)

print(accuracy_score(y_test, predRFC))
print(confusion_matrix(y_test, predRFC))
print(classification_report(y_test,predRFC))
```

```python
#Using SupportVectorClassifier

svc = SVC()
svc.fit(x_train,y_train)
predsvc = svc.predict(x_test)

print(accuracy_score(y_test, predsvc))
print(confusion_matrix(y_test, predsvc))
print(classification_report(y_test,predsvc))
```

```python
#Using LogisticRegression

LR=LogisticRegression()
LR.fit(x_train,y_train)
predLR=LR.predict(x_test)

print(accuracy_score(y_test, predLR))
print(confusion_matrix(y_test, predLR))
print(classification_report(y_test,predLR))
```

```python
#Using KNN classifier

knn = KNN()
knn.fit(x_train,y_train)
predknn = knn.predict(x_test)

print(accuracy_score(y_test, predknn))
print(confusion_matrix(y_test, predknn))
print(classification_report(y_test,predknn))
```
